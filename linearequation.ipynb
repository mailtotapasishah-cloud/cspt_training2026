```python

a = int(input("Enter a: "))
b = int(input("Enter b: "))
c = int(input("Enter c: "))

d = int(input("Enter d: "))
e = int(input("Enter e: "))
f = int(input("Enter f: "))

x = (c * e - b * f) / (a * e - b * d)
y = (a * f - c * d) / (a * e - b * d)

print("x =", x)
print("y =", y)
```
